```python
""""This dataset is composed of a range of biomedical voice measurements from 
31 people, 23 with Parkinson's disease (PD). Each column in the table is a 
particular voice measure, and each row corresponds one of 195 voice 
recording from these individuals ("name" column). The main aim of the data 
is to discriminate healthy people from those with PD, according to "status" 
column which is set to 0 for healthy and 1 for PD."""
```

```python
""""
Matrix column entries (attributes):
name - ASCII subject name and recording number
MDVP:Fo(Hz) - Average vocal fundamental frequency
MDVP:Fhi(Hz) - Maximum vocal fundamental frequency
MDVP:Flo(Hz) - Minimum vocal fundamental frequency
MDVP:Jitter(%),MDVP:Jitter(Abs),MDVP:RAP,MDVP:PPQ,Jitter:DDP - Several 
measures of variation in fundamental frequency
MDVP:Shimmer,MDVP:Shimmer(dB),Shimmer:APQ3,Shimmer:APQ5,MDVP:APQ,Shimmer:DDA - Several measures of variation in amplitude
NHR,HNR - Two measures of ratio of noise to tonal components in the voice
status - Health status of the subject (one) - Parkinson's, (zero) - healthy
RPDE,D2 - Two nonlinear dynamical complexity measures
DFA - Signal fractal scaling exponent
spread1,spread2,PPE - Three nonlinear measures of fundamental frequency variation """
```

```python
y=df['status']
x=df.drop(['status','name'],axis=1)
print(x)
```

```python
#preprocessing of data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc=sc.fit_transform(x)
print(sc)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(sc,y,test_size=0.2,random_state=123)
```

```python
#Algo 1(LOGISTIC REGRESSION)
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression().fit(x_train,y_train)
log_pred=log_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
lr_accuracy = accuracy_score(y_test, log_pred)

#print("model accuracy using logistic regression: ",accuracy_score(y_pred,y_test))
print("confusion matrix using logistic regression: ",confusion_matrix(log_pred,y_test))
print("classification report using logistic regression: ",classification_report(log_pred,y_test))
print(lr_accuracy)
```

```python
#Algo 2(GRADIENT BOOSTING ALGORITHM)
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# Train a Gradient Boosting classifier
gb_classifier = GradientBoostingClassifier()
gb_classifier.fit(x_train, y_train)
```

```python
# Make predictions on the test set
predictions = gb_classifier.predict(x_test)
```

```python
gb_accuracy = accuracy_score(y_test,predictions)

#print("model accuracy using gradient bossting: ",accuracy_score(y_pred,y_test))
print("confusion matrix using gradient boosting: ",confusion_matrix(predictions,y_test))
print("classification report using gradient boosting: ",classification_report(predictions,y_test))
print(gb_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Train a Random Forest classifier
rf_classifier = RandomForestClassifier()
rf_classifier.fit(x_train, y_train)



```

```python
# Make predictions on the test set
rf_pred = rf_classifier.predict(x_test)
```

```python
# Evaluate the model
rf_accuracy = accuracy_score(y_test,rf_pred)
print(f"Accuracy: {rf_accuracy}")
```

```python
rf_accuracy = accuracy_score(y_test,rf_pred)

#print("model accuracy using gradient bossting: ",accuracy_score(y_pred,y_test))
print("confusion matrix using gradient boosting: ",confusion_matrix(rf_pred,y_test))
print("classification report using gradient boosting: ",classification_report(rf_pred,y_test))
print(rf_accuracy)
```
